```
#Importing libraries 
from qiskit import * #Imports the whole qiskit library 
qiskit.__qiskit_version__ #checks qiskit version
import matplotlib.pyplot as plt #imports matplotlib 
%matplotlib inline 
#plots graphs after code
%config InlineBackend.figure_format = 'svg' # Makes the images look nice
import numpy as np #imports numpy
from qiskit.visualization import plot_histogram #imports plot histogram 
```

```
#Setting up quantum register
q = QuantumRegister(2,name='q')

#Setting up classical reister
c = ClassicalRegister(2,name='c')

#Setting up circuit
deutsch_qc = QuantumCircuit(q,c,name='qc')

#Initializing to 01
deutsch_qc.x(0)

#Inserting a barrier
deutsch_qc.barrier()

#Applying Hadamard Gates to each qubit
deutsch_qc.h(q)

#Inserting a barrier
deutsch_qc.barrier()

#Gates rep f_01
deutsch_qc.cx(q[1],q[0])

#Inserting a barrier
deutsch_qc.barrier()

#Applying Hadamard Gates to each qubit
deutsch_qc.h(q)

#measurement
deutsch_qc.measure(q,c)

#plot
deutsch_qc.draw(output="mpl")
```

```
#calls Aer simulator
backend = BasicAer.get_backend('qasm_simulator')

#Sets how many measurements will occur
sampleSize = 1000

#Runs Deutsch's circuit and records the outcome of each measurement
results = execute(deutsch_qc, backend=backend, shots=sampleSize).result().get_counts()

#Plots Histogram
plot_histogram(results)
```
